```python
from transformers import pipeline
```

```python
# Load BART summarizer pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# -----------------------------
# STEP 5: Input Text & Summarize
# -----------------------------

# Example long text
text = """
Artificial Intelligence (AI) has become an essential part of our lives.
It is widely used in healthcare, finance, transportation, and many other industries.
One important application of AI in Natural Language Processing (NLP) is text summarization.
Text summarization helps generate concise and meaningful summaries from large text,
making it easier to read and understand the main points quickly.
"""

# Generate summary using BART
summary = summarizer(
    text,           # Input text
    max_length=80,  # Maximum words in summary
    min_length=30,  # Minimum words in summary
    do_sample=False # Deterministic output
)

# Print results
print("Original Text:\n", text)
print("\nGenerated Summary:\n", summary[0]['summary_text'])
```
